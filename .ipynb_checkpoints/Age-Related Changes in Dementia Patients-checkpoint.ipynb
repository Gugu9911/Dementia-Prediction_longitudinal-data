```python
import pandas as pd

# Load the dataset
file_path = 'dementia_dataset.csv'
dementia_data = pd.read_csv(file_path)

# Display the first few rows of the dataset for inspection
dementia_data.head()
```

```python
# Checking for missing values
missing_values = dementia_data.isnull().sum()
missing_values
```

```python
# Converting data types if necessary
dementia_data['Age'] = dementia_data['Age'].astype(int)
dementia_data['MMSE'] = pd.to_numeric(dementia_data['MMSE'], errors='coerce')
dementia_data['nWBV'] = pd.to_numeric(dementia_data['nWBV'], errors='coerce')

# Fill or drop missing values based on your criteria
# For example, filling MMSE missing values with the mean or median
dementia_data['MMSE'].fillna(dementia_data['MMSE'].median(), inplace=True)

# Display the changes
dementia_data.head()
```

```python
# Calculating change in MMSE and nWBV for each patient across visits
dementia_data.sort_values(by=['Subject ID', 'Visit'], inplace=True)
dementia_data['MMSE_change'] = dementia_data.groupby('Subject ID')['MMSE'].diff()
dementia_data['nWBV_change'] = dementia_data.groupby('Subject ID')['nWBV'].diff()

# Grouping by age and calculating average change in MMSE and nWBV
age_group_analysis = dementia_data.groupby('Age').agg({'MMSE_change': 'mean', 'nWBV_change': 'mean'})
age_group_analysis
```

```python
# Dropping rows where MMSE_change or nWBV_change is NaN
dementia_data_cleaned = dementia_data.dropna(subset=['MMSE_change', 'nWBV_change'])

# Repeating the grouping and averaging process on the cleaned data
age_group_analysis_cleaned = dementia_data_cleaned.groupby('Age').agg({'MMSE_change': 'mean', 'nWBV_change': 'mean'})

# Display the cleaned age group analysis
age_group_analysis_cleaned
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(age_group_analysis_cleaned.index, age_group_analysis_cleaned['MMSE_change'], marker='o', linestyle='-', color='b')
plt.title('Average MMSE Change by Age Group')
plt.xlabel('Age')
plt.ylabel('Average MMSE Change')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(age_group_analysis_cleaned.index, age_group_analysis_cleaned['nWBV_change'], marker='o', linestyle='-', color='g')
plt.title('Average nWBV Change by Age Group')
plt.xlabel('Age')
plt.ylabel('Average nWBV Change')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot of MMSE Change vs. nWBV Change

plt.figure(figsize=(10, 8))
plt.scatter(dementia_data_cleaned['MMSE_change'], dementia_data_cleaned['nWBV_change'], alpha=0.5)
plt.title('Scatter Plot of MMSE Change vs. nWBV Change')
plt.xlabel('MMSE Change')
plt.ylabel('nWBV Change')
plt.grid(True)
plt.show()
```

```python
# Histogram of MMSE Changes

plt.figure(figsize=(10, 6))
plt.hist(dementia_data_cleaned['MMSE_change'], bins=20, color='skyblue')
plt.title('Distribution of MMSE Changes')
plt.xlabel('MMSE Change')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of nWBV Changes

plt.figure(figsize=(10, 6))
plt.hist(dementia_data_cleaned['nWBV_change'], bins=20, color='lightcoral')
plt.title('Distribution of nWBV Changes')
plt.xlabel('nWBV Change')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python

# Statistical tests (e.g., correlation analysis)
correlation_analysis = dementia_data[['Age', 'MMSE_change', 'nWBV_change']].corr()
correlation_analysis
```

```python
# Data Preparation for Age-Related Analysis

# Adding a column for age at each visit. Assuming that the 'Age' column represents the age at the first visit
dementia_data['Age_at_Visit'] = dementia_data['Age'] + (dementia_data['MR Delay'] / 365.25)

# Focusing on Age, MMSE, and nWBV for the analysis
age_related_data = dementia_data[['Subject ID', 'Visit', 'Age_at_Visit', 'MMSE', 'nWBV']].copy()

# Removing rows with missing MMSE or nWBV values
age_related_data.dropna(subset=['MMSE', 'nWBV'], inplace=True)

# Displaying the first few rows of the prepared data
age_related_data.head()
```

```python
missing_values_check = age_related_data.isnull().sum()
missing_values_check
```

```python
mmse_statistics = age_related_data['MMSE'].describe()
mmse_statistics
```

```python
nwbv_statistics = age_related_data['nWBV'].describe()
nwbv_statistics
```

```python
import scipy.stats as stats

# Hypothesis Test 1: Test for Differences in MMSE Scores Among Age Groups
grouped_data = age_related_data.groupby('Age_at_Visit')
age_groups = list(grouped_data.groups.keys())

# Create empty lists to store MMSE scores for each group
mmse_by_age_group = []

for age_group in age_groups:
    mmse_scores = grouped_data.get_group(age_group)['MMSE']
    mmse_by_age_group.append(mmse_scores)

# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(*mmse_by_age_group)
print("ANOVA Test Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Hypothesis Test 2: Test for Correlation Between Age and MMSE Scores
correlation_age_mmse, p_value_age_mmse = stats.pearsonr(age_related_data['Age_at_Visit'], age_related_data['MMSE'])
print("Pearson Correlation Test Results:")
print("Pearson correlation coefficient:", correlation_age_mmse)
print("p-value:", p_value_age_mmse)
```

```python
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd


# Perform one-way ANOVA to check for significant differences
# Anova test for MMSE scores among different age groups
result = stats.f_oneway(age_related_data['MMSE'], age_related_data['Age_at_Visit'])
result

# Check if the ANOVA result is significant
if result.pvalue < 0.05:  # You can adjust the significance level (alpha) as needed
    print("ANOVA result is significant. Performing Tukey's HSD test.")
    
    # Create a new DataFrame with columns 'Age_at_Visit' and 'MMSE'
    data_for_tukey = age_related_data[['Age_at_Visit', 'MMSE']]

    # Perform Tukey's HSD test
    tukey_results = pairwise_tukeyhsd(data_for_tukey['MMSE'], data_for_tukey['Age_at_Visit'])

    # Print the summary of Tukey's HSD test
    print(tukey_results.summary())
else:
    print("ANOVA result is not significant. No post hoc test needed.")
```
